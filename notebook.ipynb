```python
width = 10
height = 10
param_straight = 0.3
param_cycle = 0.05

maze = generate_maze(width, height, param_straight, param_cycle, 4)
array = maze_to_array(maze)
```

```python
# Creates a popup window in which can be drawn
%matplotlib qt
from ACORenderer import ACORenderer
from ACO import ACO

start = (1, 1)
goal = (array.shape[1] - 2, array.shape[0] - 2)
print(goal)

ants = 30
pheromones = 5.0
evaporation = 0.5
initialisation = 0.2
random_path_chance = 0.1

verbose = False

optimizer = ACO(array.copy(), start, goal, ants, pheromones, evaporation, initialisation, random_path_chance, verbose=verbose)
renderer = ACORenderer(optimizer)
renderer.show()

# after window closes
final_maze = renderer.get_maze()
```
